```python
#-----------(Import libraries)----------#
import math
import numpy as np
import matplotlib.pyplot as plt

#-----------(Heart shape using math)----------#
def hearta(k):
    return 15 * math.sin(k) ** 3

def heartb(k):
    return 12 * math.cos(k) - 6 * math.cos(2 * k) - 2 * math.cos(3 * k) - math.cos(4 * k)

#----------(Generate points for heart shape using math)----------#
x_math = [hearta(i) * 20 for i in range(9999)]
y_math = [heartb(i) * 20 for i in range(9999)]

#----------(Plotting the heart shape using math)----------#
plt.figure(figsize=(10, 5))

#----------(First subplot for math heart)----------#
plt.subplot(1, 2, 1)
plt.fill(x_math, y_math, color="#1b5e20", alpha=0.5)
plt.axis('off')
plt.axis('equal')
plt.gca().set_facecolor('white')
plt.text(0, 0, 'MSc.Chem.Tech. \n Shumookh', fontdict={'family': 'serif', 'size': 15}, ha='center', va='center', color='white')
plt.title('Heart by Math')

#----------( Heart shape using numpy)----------#
t = np.linspace(0, 2 * np.pi, 1000)
x_numpy = 15 * np.sin(t) ** 3
y_numpy = 12 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

#----------(Second subplot for numpy heart)----------#
plt.subplot(1, 2, 2)
plt.fill(x_numpy, y_numpy, color="#1b5e20", alpha=0.5)
plt.axis('off')
plt.axis('equal')
plt.gca().set_facecolor('white')
plt.text(0, 0, 'MSc.Chem.Tech. \n Shumookh', fontdict={'family': 'serif', 'size': 15}, ha='center', va='center', color='white')
plt.title('Heart by NumPy')

plt.tight_layout()
plt.show()
```
